# Parkinson Disease Prediction

Importing the Libraries

```python
import numpy as np
import pandas as pd
```

Importing the Dataset

```python
dataset = pd.read_csv("Parkinson_Disease.csv")
X= dataset.iloc[:,1:-1].values
Y=dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Dataset into Training and Test Set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
print(X_test)
```

```python
print(Y_test)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

Training the model on Training set

```python
from sklearn.svm import SVC
classifier = SVC(kernel = "rbf", random_state = 0)
classifier.fit(X_train,Y_train)
```

Predicting on the Test set

```python
Y_pred = classifier.predict(X_test)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1), Y_test.reshape(len(Y_test),1)),1))
```

Making Confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,Y_pred)
print(cm)
```

Calculating the Accuracy of the Model

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred)
```

Predicting on the new and unknown values

```python
Prediction = classifier.predict(scaler.transform([[197.07600,206.89600,192.05500,0.00289,0.00001,0.00166,0.00168,0.00498,0.01098,0.09700,0.00563,0.00680,0.00802,0.01689,0.00339,26.77500,0.422229,0.741367,-7.348300,0.177551,1.743867,0.085569]]))

if(Prediction == 1):
  print("The person has Parkinson's disease")
else:
  print("The person does not have Parkinson's disease")
```
